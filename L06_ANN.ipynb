```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt

#SÃ¦tter data op som beskrevet
x = np.linspace(-3,3,1000)
y = np.sinc(x)


plt.plot(x,y, '.')

x = x.reshape(-1,1)

mlp = MLPRegressor(activation = 'tanh',
                   hidden_layer_sizes = (5),
                   max_iter=1000,
                   solver = 'lbfgs',
                   alpha = 1e-5,
                   verbose = True)
mlp.fit(x,y)

plt.plot(x,y)
plt.plot(x, mlp.predict(x), 'rx', ms=1)
```

```python
print(mlp.coefs_) # w-parametre

print(mlp.intercepts_) # = bias le
```

```python
#%%

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.neural_network import MLPRegressor
import numpy as np


#%% data

x = np.linspace(-10,10,1000)
y = 2*np.tanh(2*x - 12) - 3*np.tanh(2*x - 4)

plt.plot(x,y, '.')

x = x.reshape(-1,1) # Scikit-algoritmer krÃ¦ver (:,1)-format


#%% fit model

mlp = MLPRegressor(activation = 'tanh', # aktiveringsfunktion 
                   hidden_layer_sizes = 2, # antal skjulte neuroner
                   alpha = 1e-5, # regulariseringsparameter, her meget lille
                   solver = 'lbfgs', # quasi-Newton solver
                   max_iter=1000,
                   verbose = True)
mlp.fit(x,y)

plt.plot(x,y)
plt.plot(x, mlp.predict(x), 'rx', ms=1)

print(mlp.coefs_) # w-parametre
print(mlp.intercepts_) # = bias led
```
